```python
# 智谱
from zhipuai import ZhipuAI
import os
api_key = os.environ.get('ZHIPUAI_API_KEY')
client = ZhipuAI(api_key=api_key) 
response = client.chat.completions.create(
    model="glm-4-plus",  
    messages=[
        {"role": "system", "content": "你是一个乐于解答各种问题的助手，你的任务是为用户提供专业、准确、有见地的建议。"},
        {"role": "user", "content": "农夫需要把狼、羊和白菜都带过河，但每次只能带一样物品，而且狼和羊不能单独相处，羊和白菜也不能单独相处，问农夫该如何过河。"}
    ],
)
print(response.choices[0].message.content)
```

```python
# deepseek
from openai import OpenAI
import os

# 从环境变量中获取API密钥
api_key = os.environ.get('DEEPSEEK_API_KEY') 
client = OpenAI(api_key=api_key, base_url="https://api.deepseek.com")

response = client.chat.completions.create(    
	model="deepseek-chat",    
	messages=[        {"role": "system", "content": "你是一个乐于助人的AI助手"},        
		           {"role": "user", "content": "你好，请为5月写一首诗，七言绝句"},   ],
    stream=False)
print(response.choices[0].message.content)
```

```python
# TJU本地模型
from openai import OpenAI
import os

client = OpenAI(
    api_key = os.environ.get('TJ_R1_KEY'),   # 'TJ_V3_KEY'
    base_url = "https://llmapi.tongji.edu.cn/v1"
)

prompt = """
请你用简洁、清晰的语言为有一定开发基础的初学者解释docker守护进程、docker desktop、容器、镜像、docker hub、kubernetes、jenkins等相关术语之间的联系
"""

chat_completion = client.chat.completions.create(
    model="DeepSeek-R1",  # 'DeepSeek-V3'，密钥也需相应替换
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    temperature = 1.0,
    top_p = 1.0,
    max_tokens = 5000
)

print(chat_completion.choices[0].message.content)
```
